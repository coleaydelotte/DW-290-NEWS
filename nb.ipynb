```python
def get_unique_values(series):
    vals = series.dropna().unique()
    return sorted(list(vals))
```

```python
# Parse Published date column
data['Published'] = pd.to_datetime(data['Published'], errors='coerce', utc=True)
data[['Published', 'URL']].head()
```

```python
# Create new date-related variables
data['Published_Year'] = data['Published'].dt.year
data['Published_Month'] = data['Published'].dt.month
data['Published_Day'] = data['Published'].dt.day
# Use normalize() to get date without time, but keep as datetime for better compatibility
data['Published_Date'] = data['Published'].dt.normalize()
data['Published_DayOfWeek'] = data['Published'].dt.day_name()
data['Published_Week'] = data['Published'].dt.isocalendar().week
data['Published_Quarter'] = data['Published'].dt.quarter

# Display new variables
data[['Published', 'Published_Year', 'Published_Month', 'Published_Day', 
      'Published_Date', 'Published_DayOfWeek', 'Published_Week', 'Published_Quarter']].head()
```

```python
# Group by dates - Articles per date
articles_by_date = data.groupby('Published_Date').agg({
    'URL': 'count',
    'Site': 'nunique',
    'Author': 'nunique',
    'Participants': 'sum',
    'RepliesCount': 'sum'
}).reset_index()
articles_by_date.columns = ['Date', 'Article_Count', 'Unique_Sites', 'Unique_Authors', 
                           'Total_Participants', 'Total_Replies']
articles_by_date = articles_by_date.sort_values('Date')
# Convert Date to date only for display
articles_by_date['Date'] = pd.to_datetime(articles_by_date['Date']).dt.date
articles_by_date.head(10)
```

```python
# Group by dates - Articles per month
articles_by_month = data.groupby(['Published_Year', 'Published_Month']).agg({
    'URL': 'count',
    'Site': 'nunique',
    'Author': 'nunique',
    'Participants': 'sum',
    'RepliesCount': 'sum'
}).reset_index()
articles_by_month.columns = ['Year', 'Month', 'Article_Count', 'Unique_Sites', 
                             'Unique_Authors', 'Total_Participants', 'Total_Replies']
articles_by_month = articles_by_month.sort_values(['Year', 'Month'])
articles_by_month
```

```python
# Group by dates - Articles per day of week
articles_by_dow = data.groupby('Published_DayOfWeek').agg({
    'URL': 'count',
    'Site': 'nunique',
    'Author': 'nunique'
}).reset_index()
articles_by_dow.columns = ['Day_of_Week', 'Article_Count', 'Unique_Sites', 'Unique_Authors']
# Order by day of week
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
# Handle NaN values in Day_of_Week
articles_by_dow = articles_by_dow[articles_by_dow['Day_of_Week'].notna()]
articles_by_dow['Day_of_Week'] = pd.Categorical(articles_by_dow['Day_of_Week'], categories=day_order, ordered=True)
articles_by_dow = articles_by_dow.sort_values('Day_of_Week')
articles_by_dow
```

```python
# Convert from long to wide form - Pivot by Site and Date
# Example: Articles per Site per Date
# Filter out rows with missing dates for pivot
data_for_pivot = data[data['Published_Date'].notna()].copy()
wide_site_date = data_for_pivot.pivot_table(
    index='Published_Date',
    columns='Site',
    values='URL',
    aggfunc='count',
    fill_value=0
)
wide_site_date.head(10)
```

```python
# Convert from long to wide form - Pivot by Language and Date
wide_language_date = data_for_pivot.pivot_table(
    index='Published_Date',
    columns='Language',
    values='URL',
    aggfunc='count',
    fill_value=0
)
wide_language_date.head(10)
```

```python
# Convert from long to wide form - Pivot by Country and Month
# Filter out rows with missing dates or countries
data_for_pivot2 = data[(data['Published_Year'].notna()) & (data['Country'].notna())].copy()
wide_country_month = data_for_pivot2.pivot_table(
    index=['Published_Year', 'Published_Month'],
    columns='Country',
    values='URL',
    aggfunc='count',
    fill_value=0
)
wide_country_month.head(10)
```

```python
# Comprehensive Missing Values Report
print("=" * 60)
print("COMPREHENSIVE MISSING VALUES REPORT")
print("=" * 60)

print("\n1. Missing Values by Column:")
print("-" * 60)
missing_by_col = data.isna().sum().sort_values(ascending=False)
missing_by_col_pct = (data.isna().sum() / len(data) * 100).sort_values(ascending=False)
missing_report = pd.DataFrame({
    'Missing_Count': missing_by_col,
    'Missing_Percentage': missing_by_col_pct.round(2)
})
missing_report = missing_report[missing_report['Missing_Count'] > 0]
print(missing_report)

print("\n2. Rows with Missing Values:")
print("-" * 60)
rows_with_missing = data[data.isna().any(axis=1)]
print(f"Total rows with at least one missing value: {len(rows_with_missing)} ({len(rows_with_missing)/len(data)*100:.2f}%)")

print("\n3. Rows with Multiple Missing Values:")
print("-" * 60)
rows_multiple_missing = data[data.isna().sum(axis=1) > 1]
print(f"Rows with more than one missing value: {len(rows_multiple_missing)} ({len(rows_multiple_missing)/len(data)*100:.2f}%)")

print("\n4. Missing Values by Date:")
print("-" * 60)
if 'Published_Date' in data.columns:
    missing_by_date = data.groupby('Published_Date').apply(lambda x: x.isna().sum().sum()).reset_index(name='Missing_Count')
    missing_by_date = missing_by_date[missing_by_date['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)
    if len(missing_by_date) > 0:
        print(missing_by_date.head(10))
    else:
        print("No missing values when grouped by date")

print("\n5. Missing Values Pattern Analysis:")
print("-" * 60)
# Check for patterns in missing values
if 'Author' in data.columns and 'Country' in data.columns:
    author_country_missing = data.groupby('Author')['Country'].apply(lambda x: x.isna().sum()).reset_index(name='Missing_Country_Count')
    author_country_missing = author_country_missing[author_country_missing['Missing_Country_Count'] > 0].sort_values('Missing_Country_Count', ascending=False)
    if len(author_country_missing) > 0:
        print("Authors with missing Country values:")
        print(author_country_missing.head(10))

print("\n6. Date Column Missing Values:")
print("-" * 60)
if 'Published' in data.columns:
    missing_dates = data['Published'].isna().sum()
    print(f"Missing Published dates: {missing_dates} ({missing_dates/len(data)*100:.2f}%)")

print("\n" + "=" * 60)
```

```python
# Create additional new variables
# Article engagement score (combination of participants and replies)
data['Engagement_Score'] = (data['Participants'].fillna(0) * 0.6 + 
                            data['RepliesCount'].fillna(0) * 0.4)

# High engagement flag
engagement_threshold = data['Engagement_Score'].quantile(0.75)
data['High_Engagement'] = data['Engagement_Score'] > engagement_threshold

# Date-based variables
data['Days_Since_Published'] = (pd.Timestamp.now(tz='UTC') - data['Published']).dt.days
data['Is_Weekend'] = data['Published_DayOfWeek'].isin(['Saturday', 'Sunday'])

# Language-English flag (handle NaN values)
data['Is_English'] = data['Language'].str.lower().fillna('') == 'english'

# Display new variables summary
print("New Variables Created:")
print(f"- Engagement_Score: {data['Engagement_Score'].describe()}")
print(f"\n- High_Engagement: {data['High_Engagement'].sum()} articles ({data['High_Engagement'].sum()/len(data)*100:.2f}%)")
print(f"\n- Is_Weekend: {data['Is_Weekend'].sum()} articles ({data['Is_Weekend'].sum()/len(data)*100:.2f}%)")
print(f"\n- Is_English: {data['Is_English'].sum()} articles ({data['Is_English'].sum()/len(data)*100:.2f}%)")

data[['Engagement_Score', 'High_Engagement', 'Days_Since_Published', 
      'Is_Weekend', 'Is_English']].head(10)
```
